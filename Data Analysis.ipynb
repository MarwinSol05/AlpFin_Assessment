# Q1. Who is the best Salesperson?

# First Approach Answer

In this approach, we assume SID as the Sales ID associated with the Sales Person. In order to find the best salesperson, we calculate the profit associated with each SID by performing the groupby operation over SID and calculating the sum wrt profit to find the best Sales Person.

```python
crm_ac.columns
```

```python
sid_profit_1 = crm_ac.groupby([' SID']).sum(' Profit').reset_index()
sid_profit_1 = sid_profit_1.sort_values(by = ' Profit', ascending = False)
print('The best salesperson is having the Sales ID:', sid_profit_1.iloc[0,:][' SID'])
```

```python
plt.figure(figsize=(15, 10));
sns.barplot(data = sid_profit_1.iloc[0:10,:], x = ' SID', y = ' Profit')
plt.xticks(rotation=45);
plt.show()
```

# Assumptions for Calculation:

1. Taking the No_of_Units_Per_Lot as 100,000 as a standard lot contains the aforementioned units of currency.
2. Taking a leverage of 10 as in order to minimize the risk
3. Taking the Rebate Amount Per Lot as \\$ 2 

```python
No_of_Units_Per_Lot = 100000
Leverage = 10
Rebate_Amount_Per_Lot = 2
```

```python
closed_trades['Nominal Volume'] = closed_trades['Volume'] * closed_trades['Open Price'] * No_of_Units_Per_Lot
closed_trades['Profit Per Volume 1mn'] = closed_trades['Profit']/(closed_trades['Nominal Volume']/1000000)
closed_trades['Margin Calculation'] = closed_trades['Volume'] * closed_trades['Open Price'] * No_of_Units_Per_Lot * (1/Leverage)
closed_trades['Rebates'] = Rebate_Amount_Per_Lot * closed_trades['Volume']
```

Renaming the Login column as Account Number such that crm_ac_cl_report and dep_withd_cl_trades report can be joined based on common Account Number such that Total Gross Profit can be obtained with respect to the Sales ID present in crm_ac_cl_report to find the best ID associated with the best salesperson.

```python
dep_withd_cl_trades.rename(columns={"Login": "Account Number"}, inplace = True)
```

```python
dat = pd.merge(crm_ac_cl_report, dep_withd_cl_trades, how = 'inner', on = 'Account Number')
dat.columns
```

```python
dat['Total Gross Profit'] = (dat['Close Price'] - dat['Open Price'] - dat[' Deposit'] + dat[' Withdraw'] - dat['Amount'])**(-1) - dat['Rebates']
dat.head()
```

# Q3. Which symbol's rpm is the best one?

We consider RPM as Risk Prioritization and Mitigation and use the least variation of Pips associated with the Symbol to determine the best category. In order to determine the least variation in Pips, we convert Pips column into its absolute form such that we can find out the symbol associated with minimum Pips.

```python
closed_trades['Symbol'].value_counts()
```

## Converting the Open Time and Close Time of the closed_trades to a date format without timestamp 

```python
closed_trades['Open Time'] = pd.to_datetime(pd.to_datetime(closed_trades['Open Time']).dt.date)
closed_trades['Close Time'] = pd.to_datetime(pd.to_datetime(closed_trades['Close Time']).dt.date)
closed_trades['Pips'] = closed_trades['Pips'].abs()
```

```python
closed_trades.head()
```

```python
sym_set = set(closed_trades['Symbol'])
sym_set
```

```python
symbols_grp = closed_trades.groupby(['Symbol'])
```

```python
sym_prof = {}

for sym in sym_set:
    
    tmp = symbols_grp.get_group(sym)
    prf = tmp['Pips'].sum()
    sym_prof[sym] = prf
```

```python
sorted_sym_prof = dict(sorted(sym_prof.items(), key=lambda item: -item[1]))
sorted_sym_prof
```

# Answer 3

```python
s = pd.Series(sorted_sym_prof)
print('The symbol with the best RPM is:',s.idxmin())
```

```python
plt.figure(figsize=(15, 10));
sns.barplot(x = s.index, y = s.values)
plt.xticks(rotation=45);
plt.show()
```
